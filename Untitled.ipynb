```python
sns.pairplot(data[['Средняя температура',
       'Минимальная температура',
       'Эффективная температура', 'Скорость ветра']])
```

```python
corr = data[['Средняя температура',
       'Минимальная температура',
       'Эффективная температура', 'Скорость ветра']].corr()
```

```python
corr
```

```python
mask = np.triu(np.ones_like(corr,dtype=bool))
cmap = sns.diverging_palette(230,20, as_cmap=True)
sns.heatmap(corr,annot=True,mask=mask, cmap=cmap)
```
